```python
from pyspark import SparkConf , SparkContext
import operator
import heapq
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

mapper1 用來處理資料並append到lines。

```python
def mapper1(line):
    wordlist = line.split(" ")
    maplist = []
    l1=[0]*58
    for i in range(58):
        l1[i] = float(wordlist[i])
    maplist.append(l1)
    return maplist
```

reducer1 用來做值得相加與list相加。

```python
def reducer1(list1,list2):
    return list1+list2
```

這裡在讀取所有的資料且做mapper1，並把c1、c2的內容先take出來，方便之後做運算及更改。

```python
conf=SparkConf().setMaster("local").setAppName("kmeans")
sc = SparkContext(conf=conf)
c1 = sc.textFile("c1.txt").flatMap(mapper1)
c2 = sc.textFile("c2.txt").flatMap(mapper1)
data = sc.textFile("data.txt").flatMap(mapper1)
c1 = c1.take(10)
c2 = c2.take(10)
```

```python
sc.stop()
```

mapper2、mapper2_1、mapper2_M1、mapper2_M2 主要在做分類，分類到距離最短的中點，並記錄是第幾個中點、自己的attributes、距離（方便後面計算cost），這四個mapper的不同在於不同的中心set（c1,c2）即不同的計算距離的方法。

```python
def mapper2(line):
    dist = 10000000
    pos = 0
    maplist = []
    for i in range(10):
        sig = 0
        for j in range(58):
            sig+=math.pow((c1[i][j]-line[j]), 2)
        sig = sig**(0.5)
        if sig<dist:
            dist = sig
            pos = i
    maplist.append([pos, line, dist**2])
    return maplist
```

```python
def mapper2_1(line):
    dist = 10000000
    pos = 0
    maplist = []
    for i in range(10):
        sig = 0
        for j in range(58):
            sig+=math.pow((c2[i][j]-line[j]), 2)
        sig = sig**(0.5)
        if sig<dist:
            dist = sig
            pos = i
    maplist.append([pos, line, dist**2])
    return maplist
```

```python
def mapper2_M1(line):
    dist = 10000000
    pos = 0
    maplist = []
    for i in range(10):
        sig = 0
        for j in range(58):
            sig+=abs(Mc1[i][j]-line[j])
        if sig<dist:
            dist = sig
            pos = i
    maplist.append([pos, line, dist])
    return maplist
```

```python
def mapper2_M2(line):
    dist = 10000000
    pos = 0
    maplist = []
    for i in range(10):
        sig = 0
        for j in range(58):
            sig+=abs(Mc2[i][j]-line[j])
        if sig<dist:
            dist = sig
            pos = i
    maplist.append([pos, line, dist])
    return maplist
```

mapper3主要在整理資料用在後面重新定義中心位置，由於重新定義中心位置是將屬於同一個中心所有點的attributes做平均，所以我們只需要
（所屬中心, attributes）。

```python
def mapper3(line):
    maplist = []
    maplist.append((line[0], line[1]))
    return maplist
```

COST 是也是在整理資料，將所有點的cost append到同一個list，並將所有的key都設為1，方便後面reduce by key。

```python
def COST(line):
    maplist = []
    maplist.append((1, line[2]))
    return maplist
```

這裡在做以c1為中心set且分類時用Euclidean distance做分類的演算。先將data做flatMap(mapper2)整理成 (所屬中心, attributes, distance)，再做flatMap(COST)整理成 (1, distance)，做reduce by key後算出每個iteration的cost並記錄起來，再用flatMap(mapper3)、reduce by key將資料整理為(中心點, 所有屬於這個中心點的attributes)，用for-loop去重新計算中心點，先算出有幾個點後將各attributes加總取平均即為新的中心點，最後開兩個二維陣列分別儲存以Euclidean及Manhattan distance算出的各中心點位置的距離，計算方法則是用雙層for-loop做公式上的運算。

最後輸出跑完20個iteration後的c1值。

```python
#Ec1
cost2 = [0]*20
for ite in range(20):
    cost = data.flatMap(mapper2)
    cost1 = cost.flatMap(COST)
    cost1 = cost1.reduceByKey(reducer1)
    cost1 = cost1.take(1)
    cost2[ite] = cost1[0][1]
    clus = cost.flatMap(mapper3)
    clus = clus.reduceByKey(reducer1)
    lines = clus.take(10)
    for line in lines:
        attr = line[1]
        num = len(line[1])/58
        num = int(num)
        for i in range(58):
            s = 0
            for j in range(num):
                s+=attr[58*j+i]
            s = s/num
            c1[line[0]][i] = s
            
print("Final value of c1:")
for i in range(10):
    print(i, c1[i])

per = (abs(cost2[19] - cost2[0]) / cost2[0])*100

Evalue=[[0]*10 for i in range(10)]
Mvalue = [[0]*10 for i in range(10)]
for i in range(10):
    for j in range(10):
        if i==j:
            Evalue[i][j]=0
            Mvalue[i][j]=0
        else:
            Esig = 0
            Msig = 0
            for k in range(58):
                Esig+=(c1[i][k]-c1[j][k])**2
                Msig+=abs(c1[i][k]-c1[j][k])
            Esig = Esig**(0.5)
            Evalue[i][j]=Esig
            Mvalue[i][j]=Msig
```
